```python
true['label']=1
```

```python
news = pd.concat([fake, true], axis=0)
```

```python
news = news.drop(['title', 'subject', 'date'], axis=1)
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'], axis=1, inplace=True)
```

```python
news.head()
```

```python
import re

```

```python
def wordopt(text):
  ## convert into lower case
    text = text.lower()
    ##Remove URLs
    text = re.sub(r'https?://\S+|www.\.\S+','',text)
    ##Remove HTML tags
    text = re.sub(r'<.*?>','',text)\
    ##Remove punctuation
    text = re.sub(r'[^\w\s]','',text)
    ##Remove digits
    text = re.sub(r'\d','',text)
    ##Remove characters
    text = re.sub(r'\n', '',text)


    return text

```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train, y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train, y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test,predict_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train, y_train)
```

```python
print(classification_report(y_test,gbc.predict(xv_test)))
```

```python
def output_lable(n):
    if n == 0:
        return "This is a Fake News"
    elif n == 1:
        return " This is not Fake News"
```

```python
def manual_testing(news):
    testing_news = {"test": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["test"] = new_def_test["test"].apply(wordopt)
    new_x_test = new_def_test["test"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    pred_dtc = DTC.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    return "LR Prediction: {} \nRFC Prediction:{} \nGBC Prediction:{}".format(
        output_lable(pred_lr[0]),
        output_lable(pred_rfc[0]),
        output_lable(pred_gbc[0])
    )
```

```python
news_article = str(input())
```

```python
print(manual_testing(news_article))
```
